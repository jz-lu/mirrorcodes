# Analysis of mirror codes
This is a high-level wrapper of the mirror code and benchmarking functionalities of this repository, enabling benchmarking of a chosen mirror code.

```python
from mirror import MirrorCode
from benchmark import StabilizerCode
import numpy as np
import matplotlib.pyplot as plt
import stim
```

As an example, let us benchmark the `[[n, k, d, w]] = [[36, 6, 6, 6]]` code.

```python
# [[36, 6, 6, 6]] code
code = MirrorCode(
        group = [2, 2, 3, 3],
        z0 = [[0, 0, 0, 0],
       [0, 1, 0, 1],
       [1, 0, 0, 2]],
        x0 = [[0, 0, 0, 0],
       [0, 1, 1, 0],
       [1, 1, 2, 0]]
    )
```

```python
print("Finding stabilizers...")
stabilizers = code.get_stim_tableau()
print("Done.")

benchmarker = StabilizerCode(stabilizers, verbose=True, name="[[36, 6, 6]]")

# Function to make the syndrome extraction circuit
# must take args of the form (p_data:float, p_meas:float, p1:float, p2:float, num_rounds:int, phenomenological:bool)
sec_maker_fn = code.barely_dressed_ancilla_sec


T = 4 # min error rate is 10^-T
NUM_PROBS = 5

# Define the main parameters of the benchmarking
ROUND_CHOICES = [3, 4]
P_DATAS = np.logspace(-T, -1, NUM_PROBS)
P_MEASS = P_DATAS / 10
P1S = P_DATAS / 10
P2S = P_DATAS / 3
PHENOMENOLOGICAL = True
NUM_SHOTS = 1000

SECS = [code.barely_dressed_ancilla_sec(p_data=P_DATAS[i], 
                                        p_meas=P_MEASS[i],
                                        p1=P1S[i],
                                        p2=P2S[i],
                                        num_rounds=nrd,
                                        phenomenological=PHENOMENOLOGICAL
                                        )
                                        for nrd in ROUND_CHOICES
                                        for i in range(len(P_DATAS))
        ]
```

```python
benchmarker.parallel_benchmark(p_datas=P_DATAS,
                               secs=SECS,
                               rounds_choices=ROUND_CHOICES,
                               num_shots=NUM_SHOTS,
                               phenomenological=PHENOMENOLOGICAL,
                               plot=True
                               )
```
