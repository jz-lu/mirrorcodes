# Analysis of mirror codes
This is a high-level wrapper of the mirror code and benchmarking functionalities of this repository, enabling benchmarking of a chosen mirror code.

```python
from mirror import MirrorCode
from benchmark import StabilizerCode, noise
import numpy as np
import matplotlib.pyplot as plt
import stim
```

As an example, let us benchmark the `[[n, k, d, w]] = [[36, 6, 6, 6]]` code.

```python
# [[36, 6, 6, 6]] code
code = MirrorCode(
        group = [2, 2, 3, 3],
        z0 = [[0, 0, 0, 0],
       [0, 1, 0, 1],
       [1, 0, 0, 2]],
        x0 = [[0, 0, 0, 0],
       [0, 1, 1, 0],
       [1, 1, 2, 0]]
    )

noise_model_name = "SI1000"
```

```python
print("Finding stabilizers...")
stabilizers = code.get_stim_tableau()
print("Done.")

benchmarker = StabilizerCode(stabilizers, verbose=False, name="[[36, 6, 6]]")

T = 6 # min error rate is 10^-T
NUM_PROBS = 8

# Define the main parameters of the benchmarking
ROUND_CHOICES = [3, 6, 9]
PS = np.logspace(-T, -1, NUM_PROBS)
# PHENOMENOLOGICAL = False
NUM_SHOTS = 10000

SECS = [code.fully_ft_for_css(noise_model_name=noise_model_name,
                              p=PS[i],
                              num_rounds=nrd
                              )
                            for nrd in ROUND_CHOICES
                            for i in range(len(PS))
        ]
```

```python
results = benchmarker.parallel_benchmark(ps=PS,
                               secs=SECS,
                               rounds_choices=ROUND_CHOICES,
                               num_shots=NUM_SHOTS,
                               phenomenological=False,
                               plot=True
                               )
```
